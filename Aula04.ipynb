```python
# Criando lista 
list = [0, 'eu amo polatm', 2.5]
print(list)

#Nesta parte fiz a adição entre duas strings, sendo que list[1] é o segundo índice do vetor list
print ('variável: ' + list [1])

# Adição do primeiro índice da váriavel list (list[0])
print(list[0]+1)
```

```python
# Criando uma matriz numpy
import numpy as np

# Criando uma matriz de zeros utilizando np.zeros
a = np.zeros((5,5))
print (a)
print('Indexação numpy')
print(a[0,0])
```

```python
# Criando números aleatórios
b=np.random.rand(5,5)
print(b)

# Mostrar a dimensão da matriz
print('Dimensão da matriz')
print(b.shape)
```

```python
# Criando matriz de valores aleatórios entre uma faixa estabelecida
vmin = 1 # Valor mínimo
vmax = 100 # Valor máximo
c = vmin+(vmax-vmin)*b
print(c)
```

```python
# Plotar matriz

import matplotlib.pyplot as plt
plt.plot(c)
```

```python
# Plotando apenas a primeira linha

plt.plot(c[0,:])
```

```python
# Plotando transposto
plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha
x=range(0,np.size(c[0,:]))
print(x)
plt.plot(c[0,:])
```

```python
#Transportando para range
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.plot(x,c[0,:])
    
```

```python
#Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,1])
ax[2].bar(x,c[:,2])

# Exportando
import os
rootFolder = os.getcwd()
print(rootFolder)
figFolder = rootFolder.replace('\\','/')+'/Figuras/'
print(figFolder)
fig.savefig('Figura01.png')                                     
```
