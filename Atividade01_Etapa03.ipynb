Gerar números aleatórios que representem a atividade industrial (ex: toneladas de cimento produzido). Utilize alguma indústria do estado para basear seus códigos. 

```python
import numpy as np

# Definindo o número de linhas e colunas da tabela
num_linhas = 5
num_colunas = 3

# Gerando dados fictícios usando NumPy
dados = np.random.randint(0, 100, size=(num_linhas, num_colunas))

# Criando os rótulos das colunas
rotulos_colunas = ["Coluna" + str(i+1) for i in range(num_colunas)]

# Imprimindo a tabela
print("Tabela de Dados Fictícios:")
# Imprimindo os rótulos das colunas
print("\t".join(rotulos_colunas))
# Imprimindo os dados
for linha in dados:
    print("\t".join(map(str, linha)))
```

```python
# Dados do eixo X e Y
eixo_x = ["janeiro", "fevereiro", "março", "abril", "maio", "junho"]
eixo_y = [38517, 37651, 36111, 38840, 41141, 75201]

# Criando uma matriz de dados da industria usando NumPy
dados = np.array([eixo_x, eixo_y]).T

# Imprimindo a tabela
print("Tabela de Dados da Industria:")
print("Eixo X\tEixo Y")
for linha in dados:
    print("\t".join(map(str, linha)))
```

```python
import matplotlib.pyplot as plt

# Criando o gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(eixo_x, eixo_y, color='skyblue')
plt.title('Gráfico de Barras')
plt.xlabel('Meses')
plt.ylabel('Valores')
plt.show()

# Criando o gráfico de linha
plt.figure(figsize=(8, 6))
plt.plot(eixo_x, eixo_y, marker='o', color='green', linestyle='-', linewidth=2)
plt.title('Gráfico de Linha')
plt.xlabel('Meses do ano de 2011')
plt.ylabel('Tonelada de Cimento Produzido em SC')
plt.grid(True)
plt.show()
```
